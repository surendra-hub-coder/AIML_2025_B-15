```python
# Step 1: Import Required Libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Step 2: Load Dataset (Using the Iris dataset as an example)
data = load_iris()
X = data.data  # Features
y = data.target  # Labels (target variable)

# Step 3: Split the Dataset into Training and Testing Sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Initialize and Train the Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)

# Step 5: Make Predictions on the Test Set
y_pred = rf_classifier.predict(X_test)

# Step 6: Evaluate the Model Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy of Random Forest Classifier: {accuracy * 100:.2f}%')
```

```python
# Step 1: Import Required Libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Step 2: Load Dataset (Using the Iris dataset as an example)
data = load_iris()
X = data.data  # Features
y = data.target  # Labels (target variable)

# Step 3: Split the Dataset into Training and Testing Sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Initialize and Train the Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)

# Step 5: Initialize and Train the Decision Tree Classifier
dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train, y_train)

# Step 6: Make Predictions on the Test Set
y_pred_rf = rf_classifier.predict(X_test)
y_pred_dt = dt_classifier.predict(X_test)

# Step 7: Evaluate the Accuracy of Both Models
accuracy_rf = accuracy_score(y_test, y_pred_rf)
accuracy_dt = accuracy_score(y_test, y_pred_dt)

# Step 8: Output the Accuracy
print(f'Accuracy of Random Forest Classifier: {accuracy_rf * 100:.2f}%')
print(f'Accuracy of Decision Tree Classifier: {accuracy_dt * 100:.2f}%')
```

```python
# Step 1: Import Required Libraries
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Step 2: Load the Iris Dataset
data = load_iris()
X = data.data  # Features (sepal length, sepal width, petal length, petal width)
y = data.target  # Labels (species of iris)

# Step 3: Split the Dataset into Training and Testing Sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Initialize and Train the Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)

# Step 5: Extract Feature Importances
feature_importances = rf_classifier.feature_importances_

# Step 6: Visualize the Feature Importances
# Get feature names
feature_names = data.feature_names

# Sort feature importance in descending order
indices = np.argsort(feature_importances)[::-1]

# Plotting the feature importances
plt.figure(figsize=(10, 6))
plt.title("Feature Importance (Random Forest)")
plt.barh(range(X.shape[1]), feature_importances[indices], align="center")
plt.yticks(range(X.shape[1]), np.array(feature_names)[indices])
plt.xlabel("Relative Importance")
plt.show()

# Step 7: Print the feature importances in sorted order
print("Feature Importance (sorted):")
for i in indices:
    print(f"{feature_names[i]}: {feature_importances[i]:.4f}")

```

```python
# Step 1: Import Required Libraries
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Step 2: Load the Iris Dataset
data = load_iris()
X = data.data  # Features (sepal length, sepal width, petal length, petal width)
y = data.target  # Labels (species of iris)

# Step 3: Split the Dataset into Training and Testing Sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Initialize and Train the Random Forest Classifier with OOB Error
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42, oob_score=True)
rf_classifier.fit(X_train, y_train)

# Step 5: Plot the OOB error over the number of trees
oob_errors = []

# Loop through different numbers of trees and record the OOB score
for n_trees in range(1, 101):
    rf_classifier.set_params(n_estimators=n_trees)
    rf_classifier.fit(X_train, y_train)
    oob_errors.append(1 - rf_classifier.oob_score_)  # 1 - OOB score gives the OOB error

# Step 6: Plot the OOB error as a function of the number of trees
plt.figure(figsize=(10, 6))
plt.plot(range(1, 101), oob_errors, marker='o', linestyle='-', color='b')
plt.title('Out-of-Bag (OOB) Error vs. Number of Trees')
plt.xlabel('Number of Trees')
plt.ylabel('OOB Error')
plt.grid(True)
plt.show()

```

```python
# Step 1: Import Required Libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, f1_score, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load the Iris Dataset
data = load_iris()
X = data.data  # Features (sepal length, sepal width, petal length, petal width)
y = data.target  # Labels (species of iris)

# Step 3: Split the Dataset into Training and Testing Sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Initialize and Train the Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)

# Step 5: Make Predictions
y_pred = rf_classifier.predict(X_test)

# Step 6: Evaluate with Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Step 7: Plot the Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=data.target_names, yticklabels=data.target_names)
plt.title("Confusion Matrix")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()

# Step 8: Calculate F1-Score
f1 = f1_score(y_test, y_pred, average='weighted')  # 'weighted' averages the F1-score for all classes
print(f'F1-score (Weighted Average): {f1:.4f}')

# Step 9: Classification Report (Precision, Recall, F1-score for each class)
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=data.target_names))
```
